```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as const
import pandas as pd
import cmath

from commpy.modulation import QAMModem, PSKModem

from optic.core import parameters
from optic.tx import simpleWDMTx
from optic.plot import pconst

from optic.models import linFiberCh, manakovSSF, ssfm, linFiberCh, awgn
from optic.models import phaseNoise, coherentReceiver, pdmCoherentReceiver, hybrid_2x4_90deg, pbs

from optic.dsp import pulseShape, firFilter, decimate, symbolSync, pnorm
from optic.equalization import edc, mimoAdaptEqualizer
from optic.carrierRecovery import cpr
from optic.metrics import fastBERcalc, monteCarloGMI, monteCarloMI, signal_power

import logging as logg
logg.getLogger().setLevel(logg.INFO)
logg.basicConfig(format='%(message)s')
```

```python
import importlib.util

# Import OFE.py (Optical Front-End)
spec = importlib.util.spec_from_file_location("OFE", "../../utils/OFE.py")
OFE = importlib.util.module_from_spec(spec)
spec.loader.exec_module(OFE)

# Import mitigationSBBI.py
spec_2 = importlib.util.spec_from_file_location("mitigationSSBI", "../../utils/mitigationSSBI.py")
SSBI = importlib.util.module_from_spec(spec_2)
spec_2.loader.exec_module(SSBI)
```

```python
# Parmetros - Transmissor
paramTx = parameters()

paramTx.M   = 64             # Ordem do formato de modulao
paramTx.constType = 'qam'   # Formato de modulao ('qam' ou 'psk')

paramTx.Rs  = 100e9         # Taxa de sinalizao [baud]
paramTx.SpS = 4             # Nmero de amostras por smbolo
Fs = paramTx.Rs*paramTx.SpS # Taxa de amostragem [Hz]

paramTx.Nbits = 300000      # Nmero total de bits por polarizao

paramTx.pulse    = 'rrc'    # Tipo do filtro formatador de pulso
paramTx.Ntaps    = 1024     # Nmero de coeficientes do filtro
paramTx.alphaRRC = 0.01     # Rolloff do pulso RRC

paramTx.Pch_dBm  = 0        # Potncia mdia por canal WDM [dBm]
paramTx.Nch      = 1        # Nmero de canais WDM

paramTx.Fc     = 193.1e12   # Frequncia central de operao
paramTx.Nmodes = 2          # Nmero de modos de polarizao

# gera sinal WDM
sigWDM_Tx, symbTx_, paramTx = simpleWDMTx(paramTx)
pconst(sigWDM_Tx)
```

```python
# Parmetros - Canal ptico
paramCh = parameters()

paramCh.Ltotal = 160       # Distncia total do enlace [km]
paramCh.alpha = 0          # Parmetro de perda da fibra [dB/km]
paramCh.D = 16             # Parmetro de disperso da fibra [ps/nm/km]
paramCh.Fc = paramTx.Fc    # Frequncia central de operao [Hz]
paramCh.Fs = Fs            # Taxa de amostragem [Hz]

sigWDM = linFiberCh(sigWDM_Tx, paramCh.Ltotal, paramCh.alpha, paramCh.D, paramCh.Fc, paramCh.Fs)
```

```python
# Potncia do sinal recebido
try:
    potRx  = 10*np.log10( 1e3 * (signal_power(sigWDM[:,0]) + signal_power(sigWDM[:,1])) )
except:
    potRx  = 10*np.log10( 1e3 * signal_power(sigWDM))
print(f'Potncia do sinal recebido: {potRx:.2f} dBm')
```

```python
# Canal AWGN
status_awgn = True  # Ativao do AWGN
OSNRdB      = 34    # OSNR do sistema

if status_awgn:
    SNRdB  = OSNRdB - 10 * np.log10( (paramTx.Rs * paramTx.Nmodes) / (2 * 12.5e9) )
    sigWDM = awgn( sigWDM, SNRdB, Fs, paramTx.Rs)

    # Potncia do sinal recebido + Potncia do rudo [dBm]
    try:
        potRx_AWGN  = 10*np.log10( 1e3 * (signal_power(sigWDM[:,0]) + signal_power(sigWDM[:,1])) )
    except:
        potRx_AWGN  = 10*np.log10( 1e3 * signal_power(sigWDM) )
    print(f'Potncia do sinal recebido + Potncia do rudo: {potRx_AWGN:.4f} dBm')
```

```python
# plot psd
fig, axs = plt.subplots(1, 2, figsize=(16,4))

axs[0].psd(sigWDM_Tx[:,0], Fs=Fs, Fc=paramTx.Fc, NFFT = 4*1024, sides='twosided', label = 'WDM spectrum - Tx')
axs[0].psd(sigWDM[:,0], Fs=Fs, Fc=paramCh.Fc, NFFT = 4*1024, sides='twosided', label = 'WDM spectrum - Rx')
axs[0].set_xlim(paramTx.Fc-Fs/2,paramTx.Fc+Fs/2)
axs[0].legend(loc='lower left')
axs[0].set_title('optical WDM spectrum - Mode 0')
axs[0].grid('True')

axs[1].psd(sigWDM_Tx[:,1], Fs=Fs, Fc=paramTx.Fc, NFFT = 4*1024, sides='twosided', label = 'WDM spectrum - Tx')
axs[1].psd(sigWDM[:,1], Fs=Fs, Fc=paramCh.Fc, NFFT = 4*1024, sides='twosided', label = 'WDM spectrum - Rx')
axs[1].set_xlim(paramTx.Fc-Fs/2,paramTx.Fc+Fs/2)
axs[1].legend(loc='lower left')
axs[1].set_title('optical WDM spectrum - Mode 1')
axs[1].grid('True')
```

```python
#############################################
# Parmetros - Receptor
mod = QAMModem(m=paramTx.M)

chIndex  = 0     # Posio do canal WDM a ser demodulado
Fc = paramTx.Fc
Ts = 1/Fs

freqGrid = paramTx.freqGrid
print(f'Demodulating channel #{chIndex} \n',
      f'- fc: {((Fc + freqGrid[chIndex])/1e12):.4f} THz\n',
      f'- : {(const.c/(Fc + freqGrid[chIndex])/1e-9):.4f} nm\n')

symbTx = symbTx_[:,:,chIndex]

#############################################
# Parmetros - Receptor - Oscilador Local
LOSPR   = 8                    # Relao potncia do sinal-LO [dB]

       = np.pi
FO      = 0*64e6                # Offset de frequncia
f_lo   = freqGrid[chIndex]+FO  # Deslocamento de frequncia do canal a ser demodulado
lw      = 0*100e3               # LineWidth
Plo_dBm = potRx + LOSPR         # Potncia [dBm]
Plo     = 10**(Plo_dBm/10)*1e-3 # Potncia [W]
_lo    = 0                     # Fase inicial [rad]    

print(f'Local oscillator\n',
      f'- Power: {Plo_dBm:.2f} dBm\n',
      f'- LOSPR: {LOSPR:.2f} dBm\n',
      f'- LineWidth: {(lw/1e3):.2f} kHz\n',
      f'- Frequency offset: {(FO/1e6):.2f} MHz')

#############################################
# Gerao do sinal LO
t       = np.arange(0, len(sigWDM))*Ts
_pn_lo = phaseNoise(lw, len(sigWDM), Ts)
sigLO   = np.sqrt(Plo)*np.exp(1j*(2**f_lo*t + _lo + _pn_lo))
```

```python
# Receptor ptico
paramRx = parameters()
paramRx.type    = 'ser'  # 'br','ser'
paramRx.algSSBI = 'dfr' # 'none', 'dfr', 'ic', 'gd'
```

```python
assert paramRx.type in ['ser', 'br'], "Topologia de receptor invlida."

if paramRx.type == 'br':
    # Receptor ptico balanceado
    paramPD = parameters()
    paramPD.B = paramTx.Rs
    paramPD.Fs = Fs    
    paramPD.ideal = True

    sig = 0 # polarization rotation angle
    sigBR = pdmCoherentReceiver(sigWDM, sigLO, sig=sig, paramPD=paramPD)

    sigRx = sigBR
    
else:
    # Receptor ptico single-ended
    paramPD = parameters()
    paramPD.B = 2 * paramTx.Rs
    paramPD.Fs = Fs
    paramPD.ideal = True

    sig = 0 # polarization rotation angle
    R1_polx, R2_polx, R1_poly, R2_poly = OFE.pdmSEReceiver(sigWDM, sigLO, sig=sig, paramPD=paramPD)
```

```python
if paramRx.type == 'ser':
    assert paramRx.algSSBI in ['none', 'dfr', 'ic', 'gd'], "Algoritmo de mitigao da SSBI invlido."
    
    if paramRx.algSSBI == 'dfr':
        # Algoritmo de reconstruo de sinal (DFR)
        Elox, Eloy = pbs(sigLO, =np.pi/4) # PBS do LO

        sigDFR_polx = SSBI.DFR(R1_polx, R2_polx, Elox)
        sigDFR_poly = SSBI.DFR(R1_poly, R2_poly, Eloy)

        sigDFR = np.array([sigDFR_polx, sigDFR_poly]).T
        sigDFR = sigDFR - np.mean(sigDFR)

        sigRx = sigDFR

    elif paramRx.algSSBI == 'ic':
        # Algoritmo iterativo de cancelamento da SSBI com clipping (IC)
        Elox, Eloy = pbs(sigLO, =np.pi/4) # PBS do LO
        
        Nsteps     = 20     # Nmero de iteraes
        enableClip = True   # Ativao o clipping

        sigIC_polx = SSBI.IC( R1_polx, R2_polx, sigWDM[:,0], Elox, N=Nsteps, clipping=enableClip)
        sigIC_poly = SSBI.IC( R1_poly, R2_poly, sigWDM[:,1], Eloy, N=Nsteps, clipping=enableClip)
        sigIC = np.array([sigIC_polx, sigIC_poly]).T

        sigRx = sigIC
        
    elif paramRx.algSSBI == 'gd':
        # Algoritmo gradiente descendente
        Elox, Eloy = pbs(sigLO, =np.pi/4) # PBS do LO
        
        # Algoritmo gradiente descendente
        mu_        = 0.05    # Passo do algoritmo
        Nsteps     = 120     # Nmero de iteraes
        enableClip = True    # Ativao o clipping

        sigGD_polx = SSBI.gradientDescent(R1_polx, R2_polx, sigWDM[:,0], Elox, mu=mu_, N=Nsteps, clipping=enableClip)
        sigGD_poly = SSBI.gradientDescent(R1_poly, R2_poly, sigWDM[:,1], Eloy, mu=mu_, N=Nsteps, clipping=enableClip)
        sigGD = np.array([sigGD_polx, sigGD_poly]).T

        sigRx = sigGD

    else:
        # No h mitigao da SSBI
        sigPD_polx = R1_polx + 1j*R2_polx
        sigPD_poly = R1_poly + 1j*R2_poly

        sigPD = np.array([sigPD_polx, sigPD_poly]).T
        sigPD = sigPD - np.mean(sigPD)

        sigRx = sigPD

# plot constellations (ps-receptor coerente)
pconst(sigRx[0::paramTx.SpS, :], lim=False, R=3, pType='fast')
```

```python
# Filtro casado
pulse = pulseShape('rrc', paramTx.SpS, N=paramTx.Ntaps, alpha=paramTx.alphaRRC, Ts=1/paramTx.Rs)
pulse = pulse/np.max(np.abs(pulse))

sigRx = firFilter(pulse, sigRx)
    
# plot constellations (ps-filtragem)
pconst(sigRx[0::paramTx.SpS,:], R=3)

# Compensao da dispero cromtica
if paramCh.Ltotal:
    sigRx = edc(sigRx, paramCh.Ltotal, paramCh.D, Fc-f_lo, Fs)
    
    # plot constellations (ps-filtragem+compensao)
    pconst(sigRx[0::paramTx.SpS,:], R=3)
```

```python
# Reduo de amostras por smbolos
paramDec = parameters()

paramDec.SpS_in  = paramTx.SpS
paramDec.SpS_out = 2

sigRx = decimate(sigRx, paramDec)
symbRx = symbolSync(sigRx, symbTx, 2)
```

```python
# Normalizao do sinal
x = sigRx
d = symbRx

x = x.reshape(len(x),2)/np.sqrt(signal_power(x))
d = d.reshape(len(d),2)/np.sqrt(signal_power(d))
```

```python
# Algoritmo de Equalizao
paramEq = parameters()
paramEq.nTaps = 15
paramEq.SpS   = 2
paramEq.mu    = [3e-3, 4e-3]
paramEq.numIter = 5
paramEq.storeCoeff = False
paramEq.alg   = ['nlms','da-rde']
paramEq.M     = paramTx.M
paramEq.L = [int(0.20*len(x)/2), int(0.80*len(x)/2)]

paramEq.prgsBar = False
y_EQ, H, errSq, Hiter = mimoAdaptEqualizer(x, dx=d, paramEq=paramEq)

# plot constellations (ps-equalizao)
discard = 5000
pconst([y_EQ[discard:-discard,:],d[discard:-discard,:]], pType='fast')
pconst(y_EQ[discard:-discard,:])
```

```python
# Algoritmo de Recuperao de Fase
paramCPR = parameters()
paramCPR.alg = 'bps'
paramCPR.M   = paramTx.M
paramCPR.N   = 160
paramCPR.B   = 64
paramCPR.pilotInd = np.arange(0, len(y_EQ), 20) 

y_CPR,  = cpr(y_EQ, symbTx=d, paramCPR=paramCPR)
y_CPR = y_CPR/np.sqrt(signal_power(y_CPR))

plt.figure(figsize=(10, 3))
plt.title('CPR estimated phase')
plt.plot(,'-')
plt.xlim(0, len())
plt.grid()

# plot constellations (ps-CPR)
discard = 5000
pconst([y_CPR[discard:-discard,:],d[discard:-discard,:]], pType='fast')
pconst(y_CPR[discard:-discard,:])
```

```python
# mtricas
for k in range(y_CPR.shape[1]):
    rot = np.mean(d[:,k]/y_CPR[:,k])
    y_CPR[:,k] = rot*y_CPR[:,k]

y_CPR = y_CPR/np.sqrt(signal_power(y_CPR))


ind = np.arange(discard, d.shape[0]-discard)
BER, SER, SNR = fastBERcalc(y_CPR[ind,:], d[ind,:], paramTx.M, 'qam')
GMI,_    = monteCarloGMI(y_CPR[ind,:], d[ind,:], paramTx.M, 'qam')
MI       = monteCarloMI(y_CPR[ind,:], d[ind,:], paramTx.M, 'qam')

polx = [BER[0],SER[0],SNR[0],MI[0],GMI[0]]
poly = [BER[1],SER[1],SNR[1],MI[1],GMI[1]]
results = np.array([polx, poly])

df2 = pd.DataFrame(results, index=['Pol. X', 'Pol. Y'], columns=['BER', 'SER', 'SNR [dB]', 'GMI [bits]', 'MI [bits]'])
df2.style.format({"BER": "{:.2e}", "SER": "{:.2e}", "SNR [dB]": "{:.2f}", "GMI [bits]": "{:.2f}", "MI [bits]": "{:.2f}"})
```
