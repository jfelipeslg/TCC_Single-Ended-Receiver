```python
import pickle
import matplotlib.pyplot as plt

# Lendo arquivos .pkl
list_M      = [4, 16, 64]            # Ordem da modulação
list_LOSPR  = [6, 8, 10, 12, 14, 16] # LOSPR do sistema

list_Ltotal  = [0]  # Distância do enlace (separando em listas)
list_SNR_0km = []
list_BER_0km = []

for idx_M in list_M:
    for idx_Ltotal in list_Ltotal:
        for idx_LOSPR in list_LOSPR:
            nameFile = f'.\\arquivosPickle\\fig3A_{idx_M}QAM_L-{idx_Ltotal}km_LOSPR-{idx_LOSPR}dB.pkl'
            with open(nameFile, 'rb') as arquivo:
                data = pickle.load(arquivo)
                list_SNR_0km.append(data['SNR'])
                list_BER_0km.append(data['BER'])
                
list_Ltotal  = [160]  # Distância do enlace (separando em listas)
list_SNR_160km = []
list_BER_160km = []

for idx_M in list_M:
    for idx_Ltotal in list_Ltotal:
        for idx_LOSPR in list_LOSPR:
            nameFile = f'.\\arquivosPickle\\fig3A_{idx_M}QAM_L-{idx_Ltotal}km_LOSPR-{idx_LOSPR}dB.pkl'
            with open(nameFile, 'rb') as arquivo:
                data = pickle.load(arquivo)
                list_SNR_160km.append(data['SNR'])
                list_BER_160km.append(data['BER'])
```

```python
plt.figure(figsize=(6,4))
plt.plot(list_LOSPR, [list_SNR_0km[y] for y in range(0,6)], linewidth=2.0, label= '4 QAM - 0km')
plt.plot(list_LOSPR, [list_SNR_0km[y] for y in range(6,12)], linewidth=2.0, label= '16 QAM - 0km')
plt.plot(list_LOSPR, [list_SNR_0km[y] for y in range(12,18)], linewidth=2.0, label= '64 QAM - 0km')
plt.plot(list_LOSPR, [list_SNR_160km[y] for y in range(0,6)], 'o-', linewidth=2.0, label= '4 QAM - 160km')
plt.plot(list_LOSPR, [list_SNR_160km[y] for y in range(6,12)], 'o-', linewidth=2.0, label= '16 QAM - 160km')
plt.plot(list_LOSPR, [list_SNR_160km[y] for y in range(12,18)], 'o-', linewidth=2.0, label= '64 QAM - 160km')
plt.ylabel('Effective SNR (dB)', fontsize=16.0, fontstyle="oblique")
plt.xlabel('LOSPR (dB)', fontsize=16.0, fontstyle="oblique")
plt.ylim(9.5,30)
plt.legend(loc='upper right', bbox_to_anchor=(1.32, 1), fontsize=8.0)
plt.grid()
plt.show()
```
