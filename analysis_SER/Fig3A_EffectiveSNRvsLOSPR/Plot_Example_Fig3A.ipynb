```python
import pickle
import matplotlib.pyplot as plt

# Lendo arquivos .pkl
list_auxM = [4, 16, 64]
list_auxLOSPR = [value for value in range(6,18,2)]

list_SNR_mode0 = [[],[]]
list_SNR_mode1 = [[],[]]

idx_Ltotal = 0 
for idx_M in list_auxM:
    for idx_LOSPR in list_auxLOSPR:
        nameFile = f'.\\arquivosPickle\\fig3A_{idx_M}QAM_L-{idx_Ltotal}km_LOSPR-{idx_LOSPR}dB.pkl'
        with open(nameFile, 'rb') as arquivo:
                data = pickle.load(arquivo)
                list_SNR_mode0[0].append(data['SNR_mode0'])
                list_SNR_mode1[0].append(data['SNR_mode1'])

idx_Ltotal = 160
for idx_M in list_auxM:
    for idx_LOSPR in list_auxLOSPR:
        nameFile = f'.\\arquivosPickle\\fig3A_{idx_M}QAM_L-{idx_Ltotal}km_LOSPR-{idx_LOSPR}dB.pkl'
        with open(nameFile, 'rb') as arquivo:
                data = pickle.load(arquivo)
                list_SNR_mode0[1].append(data['SNR_mode0'])
                list_SNR_mode1[1].append(data['SNR_mode1'])
```

```python
plt.figure(figsize=(6,4))
plt.plot(list_auxLOSPR, [list_SNR_mode0[0][y] for y in range(0,6)], '-d', linewidth=2.0, label= '4 QAM - 0km')
plt.plot(list_auxLOSPR, [list_SNR_mode0[0][y] for y in range(6,12)], '-d', linewidth=2.0, label= '16 QAM - 0km')
plt.plot(list_auxLOSPR, [list_SNR_mode0[0][y] for y in range(12,18)], '-d', linewidth=2.0, label= '64 QAM - 0km')
plt.plot(list_auxLOSPR, [list_SNR_mode0[1][y] for y in range(0,6)], '-o', linewidth=2.0, label= '4 QAM - 160km')
plt.plot(list_auxLOSPR, [list_SNR_mode0[1][y] for y in range(6,12)], '-o', linewidth=2.0, label= '16 QAM - 160km')
plt.plot(list_auxLOSPR, [list_SNR_mode0[1][y] for y in range(12,18)], '-o', linewidth=2.0, label= '64 QAM - 160km')
plt.ylabel('Effective SNR (dB)', fontsize=16.0, fontstyle="oblique")
plt.xlabel('LOSPR (dB)', fontsize=16.0, fontstyle="oblique")
plt.title('mode 0')
plt.ylim(9.5,30)
plt.legend(loc='upper right', bbox_to_anchor=(1.32, 1), fontsize=8.0)
plt.grid()
plt.show()

plt.figure(figsize=(6,4))
plt.plot(list_auxLOSPR, [list_SNR_mode1[0][y] for y in range(0,6)], linewidth=2.0, label= '4 QAM - 0km')
plt.plot(list_auxLOSPR, [list_SNR_mode1[0][y] for y in range(6,12)], linewidth=2.0, label= '16 QAM - 0km')
plt.plot(list_auxLOSPR, [list_SNR_mode1[0][y] for y in range(12,18)], linewidth=2.0, label= '64 QAM - 0km')
plt.plot(list_auxLOSPR, [list_SNR_mode1[1][y] for y in range(0,6)], 'o-', linewidth=2.0, label= '4 QAM - 160km')
plt.plot(list_auxLOSPR, [list_SNR_mode1[1][y] for y in range(6,12)], 'o-', linewidth=2.0, label= '16 QAM - 160km')
plt.plot(list_auxLOSPR, [list_SNR_mode1[1][y] for y in range(12,18)], 'o-', linewidth=2.0, label= '64 QAM - 160km')
plt.ylabel('Effective SNR (dB)', fontsize=16.0, fontstyle="oblique")
plt.xlabel('LOSPR (dB)', fontsize=16.0, fontstyle="oblique")
plt.title('mode 1')
plt.ylim(9.5,30)
plt.legend(loc='upper right', bbox_to_anchor=(1.32, 1), fontsize=8.0)
plt.grid()
plt.show()

fig, axs = plt.subplots(1, 2, figsize=(16,4))
```
