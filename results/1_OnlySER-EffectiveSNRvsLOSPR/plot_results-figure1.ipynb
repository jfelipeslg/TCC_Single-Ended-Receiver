```python
import pickle
import matplotlib.pyplot as plt

# Lendo arquivos .pkl
list_M = [4, 16, 64]
list_L = [0, 160]
list_flagAWGN = [False]
list_LOSPR = [value for value in range(6,18,2)]
list_flagRx = f'[SER-NONE]'


list_SNR = [ [[],[]], [[],[]], [[],[]] ]


for idxM, M in enumerate(list_M):
   for idxL, L in enumerate(list_L):
       for flagAWGN in [list_flagAWGN]:
           for LOSPR in list_LOSPR:
               for flagRx in [list_flagRx]:
                    nameFile = f'.\\arquivosPickle\\{M}QAM_L-{L}km_AWGN-{flagAWGN}_LOSPR-{LOSPR}dB_Receiver-{flagRx}.pkl'
                    with open(nameFile, 'rb') as arquivo:
                        data = pickle.load(arquivo)
                        list_SNR[idxM][idxL].append(data['SNR'])
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12,3))

axs[0].plot(list_LOSPR, [ list_SNR[0][0][y] for y in range(0,6)], '-d', color='blue', linewidth=2.0, label= '4 QAM - 0km')
axs[0].plot(list_LOSPR, [ list_SNR[1][0][y] for y in range(0,6)], '-d', color='orange', linewidth=2.0, label= '16 QAM - 0km')
axs[0].plot(list_LOSPR, [ list_SNR[2][0][y] for y in range(0,6)], '-d', color='green', linewidth=2.0, label= '64 QAM - 0km')
axs[0].set_ylabel('SNR (dB)', fontsize=12.0, fontstyle="normal", fontweight='bold')
axs[0].set_xlabel('LOSPR (dB)', fontsize=12.0, fontstyle="normal", fontweight='bold')
axs[0].set_ylim(9.5,30)
axs[0].legend(loc='upper left', fontsize=10.0)
axs[0].grid()

for tick in axs[0].yaxis.get_ticklabels():
    tick.set_fontweight('bold')
for tick in axs[0].xaxis.get_ticklabels():
    tick.set_fontweight('bold')

axs[1].plot(list_LOSPR, [ list_SNR[0][1][y] for y in range(0,6)], '-o', color='red', linewidth=2.0, label= '4 QAM - 160km')
axs[1].plot(list_LOSPR, [ list_SNR[1][1][y] for y in range(0,6)], '-o', color='purple', linewidth=2.0, label= '16 QAM - 160km')
axs[1].plot(list_LOSPR, [ list_SNR[2][1][y] for y in range(0,6)], '-o', color='black', linewidth=2.0, label= '64 QAM - 160km')
axs[1].set_ylabel('SNR (dB)', fontsize=12.0, fontstyle="normal", fontweight='bold')
axs[1].set_xlabel('LOSPR (dB)', fontsize=12.0, fontstyle="normal", fontweight='bold')
axs[1].set_ylim(9.5,30)
axs[1].legend(loc='upper left', fontsize=10.0)
axs[1].grid()

for tick in axs[1].yaxis.get_ticklabels():
    tick.set_fontweight('bold')
for tick in axs[1].xaxis.get_ticklabels():
    tick.set_fontweight('bold')
```
