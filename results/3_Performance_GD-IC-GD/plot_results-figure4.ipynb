```python
import pickle
import matplotlib.pyplot as plt
```

```python
list_M = [4, 16, 64]
list_auxLOSPR = [value for value in range(6,18,1)]
```

```python
SNR_SER = [ [],[],[] ]
BER_SER = [ [],[],[] ]

for idxM, M in enumerate(list_M):
    for LOSPR in list_auxLOSPR:
        nameFile = f'.\\arquivosPickle\\NONE\\{M}QAM_L-160km_AWGN-[False]_LOSPR-{LOSPR}dB_Receiver-[SER-NONE].pkl'
        with open(nameFile, 'rb') as arquivo:
                data = pickle.load(arquivo)
                SNR_SER[idxM].append(data['SNR'])
                BER_SER[idxM].append(data['BER'])
```

```python
SNR_SER_IC = [ [],[],[] ]
BER_SER_IC = [ [],[],[] ]

for idxM, M in enumerate(list_M):
    for LOSPR in list_auxLOSPR:
        nameFile = f'.\\arquivosPickle\\IC\\{M}QAM_L-160km_AWGN-[False]_LOSPR-{LOSPR}dB_Receiver-[SER-IC].pkl'
        with open(nameFile, 'rb') as arquivo:
                data = pickle.load(arquivo)
                SNR_SER_IC[idxM].append(data['SNR'])
                BER_SER_IC[idxM].append(data['BER'])
```

```python
plt.figure(figsize=(5,3))
plt.plot(list_auxLOSPR, SNR_SER[0], '-d', linewidth=2.0, label= '4 QAM')
plt.plot(list_auxLOSPR, SNR_SER[1], '-d', linewidth=2.0, label= '16 QAM')
plt.plot(list_auxLOSPR, SNR_SER[2], '-d', linewidth=2.0, label= '64 QAM')
plt.plot(list_auxLOSPR, SNR_SER_IC[0], '-o',  linewidth=2.0, label= '4 QAM w/ IC')
plt.plot(list_auxLOSPR, SNR_SER_IC[1], '-o',  linewidth=2.0, label= '16 QAM w/ IC')
plt.plot(list_auxLOSPR, SNR_SER_IC[2], '-o',  linewidth=2.0, label= '64 QAM w/ IC')
plt.ylabel('SNR (dB)', fontsize=12.0, fontstyle="normal", fontweight='bold')
plt.xlabel('LOSPR (dB)', fontsize=12.0, fontstyle="normal", fontweight='bold')
plt.xticks(fontweight='bold')
plt.yticks(fontweight='bold')
plt.grid()

# Dividir os itens da legenda em duas listas
ax = plt.gca() 
handles, labels = ax.get_legend_handles_labels()
half_len = (len(handles) + 1) // 2
handles1, labels1 = handles[:half_len], labels[:half_len]
handles2, labels2 = handles[half_len:], labels[half_len:]

# Criar duas instncias da classe Legend, uma para cada lista
legend1 = plt.legend(handles1, labels1, loc='lower right', fontsize=8.5)
legend2 = plt.legend(handles2, labels2, loc='upper left', fontsize=8.5)

# Adicionar as instncias de Legend ao grfico
ax.add_artist(legend1)
ax.add_artist(legend2)
```
