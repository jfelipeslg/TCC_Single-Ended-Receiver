```python
import numpy as np
import matplotlib.pyplot as plt
import cmath
```

```python
## Clipping de sinais reais
def clippingReal(signal, x_min, x_max):
    
    sigClipped = np.zeros(signal.shape)
    
    for idx in range(len(signal)):
              
        if signal[idx] < x_min:
            sigClipped[idx] = x_min
        
        elif x_max < signal[idx]:
            sigClipped[idx] = x_max
        
        else:
            sigClipped[idx] = signal[idx]
            
    return sigClipped
```

```python
sig1_real = np.sin( np.linspace(0, 10, num = 500) )
x_min = -0.70 * max(abs(sig1_real))
x_max =  0.70 * max(abs(sig1_real))
print(f'Valor máximo do sinal 1: { max(abs(sig1_real)) } | x_min = { x_min } | x_max = { x_max }')
sig1_real_clipped = clippingReal(sig1_real, x_min=x_min, x_max=x_max)

sig2_real = np.random.poisson(5, 50)
x_min = -0.70 * max(abs(sig2_real))
x_max =  0.70 * max(abs(sig2_real))
print(f'Valor máximo do sinal 2: { max(abs(sig2_real)) } | x_min = { x_min } | x_max = { x_max }')
sig2_real_clipped = clippingReal(sig2_real, x_min=x_min, x_max=x_max)

sig3_real = 2 * np.random.randint(2, size=50) - 1
x_min = -0.70 * max(abs(sig3_real))
x_max =  0.70 * max(abs(sig3_real))
print(f'Valor máximo do sinal 3: { max(abs(sig3_real)) } | x_min = { x_min } | x_max = { x_max }')
sig3_real_clipped = clippingReal(sig3_real, x_min=x_min, x_max=x_max)
```

```python
# plot figures
fig, axs = plt.subplots(3, 2, figsize=(16,9))

# sig1
axs[0][0].plot(sig1_real.real, 'r-')
axs[0][0].plot(sig1_real_clipped.real, 'k-')
axs[0][0].set_title('real')

axs[0][1].plot(sig1_real.imag, 'r-')
axs[0][1].plot(sig1_real_clipped.imag, 'k-')
axs[0][1].set_title('imag')

# sig2
axs[1][0].plot(sig2_real.real, 'r-')
axs[1][0].plot(sig2_real_clipped.real, 'k-')
axs[1][0].set_title('real')

axs[1][1].plot(sig2_real.imag, 'r-')
axs[1][1].plot(sig2_real_clipped.imag, 'k-')
axs[1][1].set_title('imag')

# sig3
axs[2][0].plot(sig3_real.real, 'r-')
axs[2][0].plot(sig3_real_clipped.real, 'k-')
axs[2][0].set_title('real')

axs[2][1].plot(sig3_real.imag, 'r-')
axs[2][1].plot(sig3_real_clipped.imag, 'k-')
axs[2][1].set_title('imag')
```

```python
## Clipping de sinais complexos
def clippingComplex(signal, clippingValue):
    
    sigClipped = np.zeros(signal.shape, dtype="complex")
    
    for idx, idxValue in enumerate(signal):
        
        if np.abs(idxValue) > clippingValue:
            absIdx, angIdx = cmath.polar(idxValue)
            sigClipped[idx] = cmath.rect(clippingValue, angIdx)
            
        else:
            sigClipped[idx] = idxValue

    return sigClipped
```

```python
sig1_complex = np.sin( np.linspace(0, 10, num = 500) ) + 1j*np.sin( np.linspace(0, 10, num = 500) )
clippingValue = 0.70 * max(abs(sig1_complex))
print(f'Valor máximo do sinal 1: {max(abs(sig1_complex))} | clippingValue: {clippingValue}')
sig1_complex_clipped = clippingComplex(sig1_complex, clippingValue=clippingValue)

sig2_complex = np.random.poisson(5, 50) + 1j*np.random.poisson(5, 50)
clippingValue = 0.70 * max(abs(sig2_complex))
print(f'Valor máximo do sinal 2: {max(abs(sig2_complex))} | clippingValue: {clippingValue}')
sig2_complex_clipped = clippingComplex(sig2_complex, clippingValue=clippingValue)

sig3_complex = (2 * np.random.randint(2, size=50) - 1) + 1j*(2 * np.random.randint(2, size=50) - 1)
clippingValue = 0.70 * max(abs(sig3_complex))
print(f'Valor máximo do sinal 3: {max(abs(sig3_complex))} | clippingValue: {clippingValue}')
sig3_complex_clipped = clippingComplex(sig3_complex, clippingValue=clippingValue)
```

```python
# plot figures
fig, axs = plt.subplots(3, 3, figsize=(16,10))

# sig1
axs[0][0].plot(sig1_complex.real, 'r-')
axs[0][0].plot(sig1_complex_clipped.real, 'k-')
axs[0][0].set_title('real')

axs[0][1].plot(sig1_complex.imag, 'r-')
axs[0][1].plot(sig1_complex_clipped.imag, 'k-')
axs[0][1].set_title('imag')

axs[0][2].plot(abs(sig1_complex), 'r-')
axs[0][2].plot(abs(sig1_complex_clipped), 'k-')
axs[0][2].set_title('abs')

# sig2
axs[1][0].plot(sig2_complex.real, 'r-')
axs[1][0].plot(sig2_complex_clipped.real, 'k-')
axs[1][0].set_title('real')

axs[1][1].plot(sig2_complex.imag, 'r-')
axs[1][1].plot(sig2_complex_clipped.imag, 'k-')
axs[1][1].set_title('imag')

axs[1][2].plot(abs(sig2_complex), 'r-')
axs[1][2].plot(abs(sig2_complex_clipped), 'k-')
axs[1][2].set_title('abs')

# sig3
axs[2][0].plot(sig3_complex.real, 'r-')
axs[2][0].plot(sig3_complex_clipped.real, 'k-')
axs[2][0].set_title('real')

axs[2][1].plot(sig3_complex.imag, 'r-')
axs[2][1].plot(sig3_complex_clipped.imag, 'k-')
axs[2][1].set_title('imag')

axs[2][2].plot(abs(sig3_complex), 'r-')
axs[2][2].plot(abs(sig3_complex_clipped), 'k-')
axs[2][2].set_title('abs')
```
